```python
# Elabore um programa que exiba na tela os números de 1 até 20, um em cada linha.
contador = 0
while contador < 20:
    contador += 1
    print(contador)
```

```python
# Solicite ao usuário um número inteiro positivo. Utilizando um laço de repetição, calcule e exiba a soma de todos os números de 1 até o número informado
n = int(input("Digite um número: "))

soma = 0
for i in range(n + 1):
    soma += i

print(f"A soma dos números de 0 até {n} é: {soma}")
```

```python
# Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.
numero_tabuada = int(input("Digite um número: "))

for i in range(1, 11):
    resultado = numero_tabuada * i
    print(f"{numero_tabuada} x {i} = {resultado}")

```

```python
# Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.
numb = int  (input("Digite um numero inteiro positivo: "))
while numb < 0:
  numb = int  (input("Por favor, digite um numero inteiro positivo: "))
fatorial = 1
for i in range (1,numb+1):
  fatorial = fatorial * i
print(f'Fatorial = {fatorial}')
print('Pois:')
print(f'{numb}! = ',end='')
for i in range (numb,1,-1):
  print(f'{i} × ',end='')
print('1')
```

```python
# Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.
valor_saque = int(input("Digite o valor do saque: "))

notas = [100, 50, 20, 10, 5, 2, 1]

print("\nSaída esperada:\n")

for nota in notas:
    quantidade = valor_saque // nota
    if quantidade > 0:
        print(f"{quantidade} nota(s) de R${nota}")
    valor_saque %= nota
```
