```python
# 2 - conhtagem regressiva, o usuario digita um numero e o prograqma imprimi do numero até o zero

n = int(input("Digite um número: "))

for i in range(n, -1, -1):
    print(i)
```

```python
# 3 - numeros pares, o usuario digita um numero e o programa imprimi os numeros e o programa imprimi os numeros pares

n = int(input("Digite um número: "))

for i in range(0, n + 1):
    if i % 2 == 0:
        print(i)
```

```python
# 4 - numeros impares,o usuario digita um numero e o programa imprimi os numeros impares

n = int(input("Digite um número: "))

for i in range (0, n + 1):
    if i % 2 != 0:
        print(i)
```

```python
# 5 - soma, o usuario digita um numero e o programa soma os numeros do zero até o numero

n = int(input("Digite um número: "))

soma = 0
for i in range(n + 1):
    soma += i

print(f"A soma dos números de 0 até {n} é: {soma}")
```

```python
# 6- tabuada, o usuario digita um numero e o programa imprimi a tabuada do numero

a = int(input("Digite um número para ver a tabuada: "))

for i in range(1, 11):
    print(f"{a} x {i} = {a * i}")
```

```python
 # 7 - contar letras, o usuario digita uma palavra e o programa conta as letras da palavra

 palavra = input("Digite uma palavra: ")
quantidade_letras = len(palavra)
print(f"A palavra '{palavra}' tem {quantidade_letras} letras. ")
```

```python
# 8 - contar vogais, o usuario digita uma palavra e oprograma conta as vogais

palavra = input("Digite uma palavra: ")
vogais = "aeiouAEIOU"
contador_vogais = 0

for letra in palavra:
    if letra in vogais:
        contador_vogais += 1
        print(f"a palavra  '{palavra}' tem {contador_vogais}vogais")
```

```python
# 9 - maior número digitado, o usuário digita quatro números e o programa imprimi o maior número digitado
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
num3 = int(input("Digite o terceiro número: "))
num4 = int(input("Digite o quarto número: "))

maior_numero = max(num1, num2, num3, num4)
print(f"O maior número digitado é: {maior_numero}")
```

```python
# 10 - o usuário digita um número entre 01 e 100 e o programa mostra quais são divisiveis por 3 e 5
numero_limite = int(input("Digite um número entre 1 e 100: "))

if 1 <= numero_limite <= 100:
    print(f"Números divisíveis por 3 e 5 até {numero_limite}:")
    for i in range(1, numero_limite + 1):
        if i % 3 == 0 and i % 5 == 0:
            print(i)
else:
    print("Por favor, digite um número entre 1 e 100.")
```
